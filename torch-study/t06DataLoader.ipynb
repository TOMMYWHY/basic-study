```python
import numpy as np
import pandas as pd
import sklearn
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
class DiabetesDataset(Dataset):
    def __init__(self,filepath):
        df= pd.read_csv(filepath,dtype=np.float32)
        self.len=df.shape[0]
        self.x_data=torch.from_numpy(np.array(df.iloc[:,:-1]))
        self.y_data=torch.from_numpy(np.array(df.iloc[:,-1]))

    
    def __getitem__(self,index):
        return self.x_data[index],self.y_data[index]
    
    
    def __len__(self):
        return self.len
    
dataset= DiabetesDataset('./data/diabetes.csv.gz')
train_loader=DataLoader(dataset=dataset,batch_size=32,shuffle=True,num_workers=2)
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1 = torch.nn.Linear(8,6,True) # input output
        self.linear2 = torch.nn.Linear(6,4,True) # input output
        self.linear3 = torch.nn.Linear(4,1,True) # input output
        self.sigmoid=torch.nn.Sigmoid()
        
        
    def forward(self,x):
        """__call__ 中被调用  """
        x=self.sigmoid(self.linear1(x))
        x=self.sigmoid(self.linear2(x))
        x=self.sigmoid(self.linear3(x))
        return x
    
model = Model()

# criterion =torch.nn.MSELoss(size_average=False)
criterion =torch.nn.BCELoss(size_average=False) # 交叉熵
optimizer=torch.optim.SGD(model.parameters(),lr=0.1)
```

```python
for epoch in range(100):
    for i,data in enumerate(train_loader,0):
        inputs,labels=data
        y_pred =model(inputs)
        loss=criterion(y_pred,labels)
#         if i% 10==0:
#             print(epoch,i,loss.item())

        print(epoch,i,loss.item()) # todo?? i max=24???

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
```
