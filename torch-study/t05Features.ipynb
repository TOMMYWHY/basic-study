```python
import numpy as np
import pandas as pd
import sklearn
import torch
```

```python
df = pd.read_csv("./data/diabetes.csv.gz")
```

```python
df
```

```python
x_data=df.iloc[:,:-1]
y_data=df.iloc[:,-1]
```

```python
x_data_tensor=torch.from_numpy(np.array(x_data))
y_data_tensor =torch.from_numpy(np.array(y_data))
```

```python
x_data_tensor
# y_data_tensor
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1 = torch.nn.Linear(8,6,True) # input output
        self.linear2 = torch.nn.Linear(6,4,True) # input output
        self.linear3 = torch.nn.Linear(4,1,True) # input output
        self.sigmoid=torch.nn.Sigmoid()
        
        
    def forward(self,x):
        """__call__ 中被调用  """
        x=self.sigmoid(self.linear1(x))
        x=self.sigmoid(self.linear2(x))
        x=self.sigmoid(self.linear3(x))
        return x
    
model = Model()

# criterion =torch.nn.MSELoss(size_average=False)
criterion =torch.nn.BCELoss(size_average=False) # 交叉熵
optimizer=torch.optim.SGD(model.parameters(),lr=0.1)
```

```python
for epoch in range(1000):
    y_pred = model(x_data_tensor.float())
    loss = criterion(y_pred, y_data_tensor.float())
    print(epoch,loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
# print("w=",model.linear.weight.item())
# print("b=",model.linear.bias.item())
```
